```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
import math
```

## 1

#### 1.1

```python
data = pd.read_csv('2.csv', header = None, names=list('XYZK'))
```

#### 1.2

```python
x = data['X']
y = data['Y']
z = data['Z']
k = data['K']
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(x, y, z, c = k, s = 50)
```

#### 1.3

```python
del data['K']
```

```python
pca = PCA(n_components=3)
A = pca.fit_transform(data)
```

#### 1.4

```python
print(pca.explained_variance_ratio_)
```

#### 1.5

```python
A = pd.DataFrame(data = A)
a1 = A[0]
a2 = A[1]
```

```python
fig, ax = plt.subplots()
ax.scatter(a1, a2, c = k, s = 50)
```

## 3

#### 3.1

```python
data3 = pd.read_csv('222.csv', header = None, names = list('ABCK'))
```

#### 3.2

```python
x3 = data3['A']-50
y3 = data3['B']-50
z3 = data3['C']+50
k3 = data3['K']
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(x3, y3, z3, c = k3, s = 50)
```

#### 3.3(    )

```python
r = (x3**2+z3**2+y3**2)**(1/2)
o = [math.atan(math.sqrt(x3[i]**2 + y3[i]**2)/z3[i]) for i in range(0, x3.size)]
f = [math.atan(y3[i]/x3[i]) for i in range(0, x3.size)]
XXX = pd.DataFrame({'A' : r, 'B' : o, 'C' : f})
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(r, o, f, c = k3, s = 10)
```

#### 3.4

```python
pca3 = PCA()
XXX = pca3.fit_transform(XXX)
print(pca3.explained_variance_ratio_)
```

#### 3.5

```python
XXX = pd.DataFrame(data = XXX)
r = XXX[0]
o = XXX[1]
```

```python
fig, ax = plt.subplots()
ax.scatter(r, o, c = k3, s = 50)
```
